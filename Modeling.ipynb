# Подготовка данных для обучения YOLO

```python
import pandas as pd
import shutil
import os
from PIL import Image
```

```python
df_val = pd.read_csv('UECFOOD256/test_list.csv', delimiter=" ", header=0)
df_train = pd.read_csv('UECFOOD256/train_list.csv', delimiter=" ", header=0)
df_test = pd.read_csv('UECFOOD256/val_list.csv', delimiter=" ", header=0)
```

```python
df_val['filename'] = df_val.apply(lambda x: x['path_img'].split('/')[2], axis=1)
df_train['filename'] = df_train.apply(lambda x: x['path_img'].split('/')[2], axis=1)
df_test['filename'] = df_test.apply(lambda x: x['path_img'].split('/')[2], axis=1)
```

## Данные validation

```python
PATH_s_f = r"C:\Users\krugd\OneDrive\Рабочий стол\HSE_progect\dataset\AIYP24-Calorie-Tracker\UECFOOD256"
PATH_t_i = r"C:\Users\krugd\OneDrive\Рабочий стол\HSE_progect\dataset\AIYP24-Calorie-Tracker\data_for_YOLO\val\images"
PATH_t_l = r"C:\Users\krugd\OneDrive\Рабочий стол\HSE_progect\dataset\AIYP24-Calorie-Tracker\data_for_YOLO\val\labels"
```

```python
# Папка, куда будут перенесены изображения
images_folder = PATH_t_i
# Папка, куда будут помещены текстовые файлы
labels_folder = PATH_t_l

# Перенос файлов и создание меток
for index, row in df_val.iterrows():
    filename = row['filename']
    category_id = row['category_id']
    x1 = row['x1']
    y1 = row['y1']
    x2 = row['x2']
    y2 = row['y2']

    # Путь к исходным изображениями
    i = 1
    for i in range(256):
        source_image_path = f'UECFOOD256/{i}/{filename}'
        if os.path.exists(source_image_path):
            # Копируем изображение в папку images
            shutil.copy(source_image_path, os.path.join(images_folder, filename))
            break
    if os.path.exists(source_image_path):
        with Image.open(source_image_path) as img:
    # Получаем размер изображения
            width, height = img.size

        x1_new = x1/width
        y1_new = y1/height
        x2_new = x2/width
        y2_new = y2/height

        width_bb = abs(x2_new - x1_new)
        height_bb = abs(y2_new - y1_new)
        x_center = x1_new + width_bb / 2
        y_center = y1_new + height_bb / 2

    # Создаем соответствующий txt файл с меткой
    label_file_path = os.path.join(labels_folder, f"{os.path.splitext(filename)[0]}.txt")
    if os.path.isfile(label_file_path):
        with open(label_file_path, 'a', encoding='utf-8') as file:
            file.write(f"{category_id-1} {x_center} {y_center} {width_bb} {height_bb}\n")
    else:
        with open(label_file_path, 'w') as f:
            f.write(f"{category_id-1} {x_center} {y_center} {width_bb} {height_bb}\n")

print("Перенос изображений и создание меток завершено.")
```

```python
def count_files_in_directory(directory):
    # Получаем список файлов и папок в указанной директории
    files = os.listdir(directory)
    # Считаем только файлы
    file_count = sum(1 for f in files if os.path.isfile(os.path.join(directory, f)))
    return file_count
```

# Обучение YOLO

```python
from ultralytics import YOLO
```

### Load YOLO11s model

```python
model_11 = YOLO("yolo11s.pt")
```

### Train YOLO11s

```python
PATH_data_yaml = r"C:\Users\krugd\OneDrive\Рабочий стол\HSE_progect\dataset\AIYP24-Calorie-Tracker\data.yaml"
```

```python
results_11s_3 = model_11.train(
    data = PATH_data_yaml, 
    epochs = 100, 
    imgsz = 480,
    batch = 32,
    lr0 = 0.0015,
    device = "0",
    name = 'yolov11s_v3'
)
```

```python
model_11.save('models/custom_yolov11s_e100.pt')
```

```python
im = r"C:\Users\krugd\OneDrive\Рабочий стол\HSE_progect\dataset\AIYP24-Calorie-Tracker\data_for_YOLO\test\images"
```

```python
results = model_11(im)
```

```python
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    result.show()  # display to screen
    result.save('test_result/result_2.jpg')  # save to disk
```
